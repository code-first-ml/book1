```python
import jax 
import jax.numpy as jnp
import tensorflow_probability.substrates.jax as tfp
import arviz as az
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
gamma = tfp.distributions.Gamma(1, 1)
beta = tfp.distributions.Beta(2, 2)
```

```python
%%capture
gamma_data = gamma.sample(seed = jax.random.PRNGKey(0), sample_shape=(1000))
beta_data = beta.sample(seed = jax.random.PRNGKey(0), sample_shape=(1000))
```

```python
sns.kdeplot(gamma_data, label='Seaborn', color='C0')
az.plot_dist(gamma_data, label='Arviz', color='C1')
plt.title("Gamma")
sns.despine()
```

```python
sns.kdeplot(beta_data, label='Seaborn', color='C0')
sns.kdeplot(beta_data, label='Seaborn-Clipped', color='C2', clip=(0.0, 1.0), lw=4, alpha=1, ls='--')
az.plot_dist(beta_data, label='Arviz', color='C1')
plt.title("Beta")
sns.despine()
```
